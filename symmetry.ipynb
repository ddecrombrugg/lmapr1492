```python
from pymatgen.ext.matproj import MPRester
from IPython.display import Image
from pymatgen.core.operations import SymmOp
from pprint import pprint
with MPRester("B6FsiVusbpLKdMu3") as m: 
    structure = m.get_structure_by_material_id("mp-30031")
```

```python
pprint("Premier élement de symétrie x-y, -y, -z")
E1 = SymmOp.from_xyz_string("x-y, -y, -z")
pprint("Matrice de rotation : ")
pprint(E1.rotation_matrix)
pprint("Vecteur de translation : ")
pprint(E1.translation_vector)
pprint("Coordonées initiales : [0.0,  0.47,  0.67]")
pprint("Coordonées finales : ")
pprint(E1.operate([0.0,  0.47,  0.67]))

Image(filename='x-y, -y, -z.png')
```
